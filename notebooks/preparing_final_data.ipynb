### Algerian written in latin data

Algerian-written-in-latin data is stored in many xml files that exist them selves in sub folders of sub folders of Datasets/text_emotion_algeria_latin folder. That means we should read them iteratively, then structure and store them in one dataframe.

```python
#Defining a loop to read xml files
#Initializing the alg_data
alg_data = { 'original' : ['First row to be deleted'],
             'sentiment': ['First row to be deleted']}
alg_data = pd.DataFrame ( alg_data)
#The head folder path
head_path = 'D:/Users/chatbot_emotions_detection/Datasets/text_emotion_algeria_latin'
folders_1 = os.listdir (head_path)
#First level path ()
for folder_1 in folders_1:
    xml_file_path = os.path.join ( head_path, folder_1 ) 
    #print ( '1 : xml file path:', xml_file_path )
    folders_1_1 = os.listdir(xml_file_path) 
    #Second level path(femal, male)
    for folder_1_1 in folders_1_1:
        xml_file_path_1 = os.path.join ( xml_file_path, folder_1_1 ) 
        #print ( '2 : xml file path:', xml_file_path_1 )
        folders_1_1_1 = os.listdir (xml_file_path_1)
        #Third level path(F_1.xml, ...)
        for folder_1_1_1 in folders_1_1_1:
            #XML file's final path
            XML_FILE_PATH = os.path.join ( xml_file_path_1, folder_1_1_1 )
            #print ( '3 : xml file path:', XML_FILE_PATH )
            #Reading the file
            new_row = pd.read_xml (XML_FILE_PATH, xpath='//doc')
            #Adding the file to our data
            alg_data = pd.concat ( [alg_data, new_row[['original', 'sentiment']]] )
```

```python
#reinitializing the index 
alg_data = alg_data.reset_index(drop = True)
#removing the first row
alg_data = alg_data.drop (index = 0)
#Adding a new column [language]
alg_data ['language']= 'alg-latin'

#Renaming alg_data column to text emotion and language
alg_data.columns = ['text', 'emotion', 'language']
#Alg data is ready
alg_data
```

```python
#Defining delete_username function
def delete_username (sentence):
    """
    This function takes as input a sentence and delete all the words in it that start with the caracter @
    sentence: str
    exemple : '@Hamza I don't agree with this politic.' ----> 'I don't agree with this politic'
    """
    #Get tokens from sentence
    tokens = nltk.word_tokenize (sentence)
    #initiate a list that will contains the remaning tokens
    final_tokens = []
    #Loop over these tokens
    i = 0
    while i < len (tokens):
        #See if the token is starting with a @
        if tokens[i].startswith("@"):
            #if so skip it and the next word
            i = i+2
        else:
            #The word doesn't start with @ so add it
            final_tokens.append (tokens[i])
            i = i+1

    #Finally concatenate tokens to a sentence
    sentence = ' '.join (final_tokens)
    return sentence

#Testing the function
sentence = 'Hello @Hamza how are you doing ?, did you see @mohamed I want to talk with him'
print ("sentence:",  sentence)
print ("sentence after deleting usernames :", delete_username(sentence))
```

```python
#Defining delete_username function
def delete_links (sentence):
    """
    This function takes as input a sentence and delete all the links in it that start with the caracter http
    sentence: str
    Link exemple : "https : //t.co/LiVWCUPwxY"
    """
    #Get tokens from sentence
    tokens = nltk.word_tokenize (sentence)
    #initiate a list that will contains the remaning tokens
    final_tokens = []
    #Loop over these tokens
    i = 0
    while i < len (tokens):
        #See if the token is starting with a @
        if tokens[i].startswith("http"):
            #if so skip it and the next word
            i = i+3
        else:
            #The word doesn't start with @ so add it
            final_tokens.append (tokens[i])
            i = i+1

    #Finally concatenate tokens to a sentence
    sentence = ' '.join (final_tokens)
    return sentence
```
